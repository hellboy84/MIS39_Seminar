```
時系列データの長期的な傾向を視覚化するには、
移動平均を使用すると良いでしょう。

移動平均は、短期的な変動を平滑化し、

特に、12ヶ月移動平均（年間平均）は、
季節性を考慮したデータの長期トレンドを捉えるのに適しています。

以下のRコードは、全入館者数に対する12ヶ月移動平均を計算し、
それをグラフ上に描画する方法を示しています。
```

````r
# パッケージの読み込み
install.packages("zoo")
library(zoo)

# 12ヶ月移動平均の計算
data$MovingAverage <- rollmean(data$all, 12, fill = NA, align = 'right')

# 元の時系列データと移動平均を用いたライングラフの描画
ggplot(data, aes(x = Date)) +
    geom_line(aes(y = all), colour = "blue", alpha = 0.5) +  # 元のデータ
    geom_line(aes(y = MovingAverage), colour = "red") +  # 移動平均
    labs(title = "Monthly Total Visitors to the Library with 12-Month Moving Average",
         x = "Date", y = "Number of Visitors") +
    theme_minimal()
````

```
季節変動や周期性を視覚的に確認する一つの方法は、
季節分解を使用することです。

これは、時系列データをトレンド、季節成分、そして残差成分に分ける手法です。
R言語では、stl 関数を使用してこの分解を行うことができます。
これにより、元のデータから季節成分を抽出し、
それを別々に表示することが可能です。

以下のコードでは、
stl 関数を用いて季節分解を行い、結果をグラフで表示します。

ここで、データが月次データであることを考慮して、
s.window オプションを "periodic" に設定しています。
これにより、全ての年で同じ季節性パターンが繰り返されると仮定します。
```




````r
# パッケージの読み込み
install.packages("tseries")
install.packages("forecast")
library(tseries)
library(forecast)

# 時系列オブジェクトの作成
ts_data <- ts(data$all, start = c(2012, 4), frequency = 12)

# 季節分解 (STL)
decomposed <- stl(ts_data, s.window = "periodic")

# 分解結果のプロット
plot(decomposed)
````
